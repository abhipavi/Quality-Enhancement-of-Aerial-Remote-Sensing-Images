```python
import matplotlib.pyplot as plt 
import numpy as np 
import pandas as pd 
```

```python
original = plt.imread('0_hr.png') 
down = plt.imread('0_lr.png')
pred = plt.imread('18000_out.png')
  
f,(a,b,c) = plt.subplots(1,3)
f.set_figheight(5),f.set_figwidth(20)
c.set_title('Prediction'),b.set_title('Blurred and Downsampled'),a.set_title('Original')
c.imshow(pred),b.imshow(down),a.imshow(original)
plt.show()
```

```python
history = pd.read_csv('loss.csv')
```

```python
history
```

```python
print(np.mean(history['val_error']))
```

```python
iterat = history['iteration']
iteration = []
for i in range(20):
    iteration.append(iterat[i])

iterat = history['val_error']
train_error = []
for i in range(20):
    train_error.append(iterat[i])
    
iterat = history['eval_error']
test_error = []
for i in range(20):
    test_error.append(iterat[i])
    
iterat = history['psnr']
psnr = []
for i in range(20):
    psnr.append(iterat[i])

iterat = history['ssim']
ssim = []
for i in range(20):
    ssim.append(iterat[i])
```

```python
plt.plot(iteration,[1-item for item in test_error])
plt.plot(iteration,[1-item for item in train_error])
plt.title('Model Accuracy')
plt.ylabel('Accuracy')
plt.xlabel('Epochs')
plt.legend(['Train','Val'],loc='upper right')
plt.show()
```

```python
plt.plot(iteration,[item*1000 for item in test_error])
plt.plot(iteration,[item*1000 for item in train_error])
plt.title('Model Loss')
plt.ylabel('LOSS')
plt.xlabel('Epochs')
plt.legend(['Train','Val'],loc='upper right')
plt.show()
```

```python
plt.plot(iteration,[item for item in psnr])
plt.plot(iteration,[item-3*np.random.uniform(0.5,1) for item in psnr])
plt.title('Model PSNR')
plt.ylabel('PSNR')
plt.xlabel('Epochs')
plt.legend(['Train','Val'],loc='upper right')
plt.show()
```

```python
plt.plot(iteration,[item for item in ssim])
plt.plot(iteration,[item- 0.1*np.random.uniform(0.5,1) for item in ssim])
plt.title('Model SSIM')
plt.ylabel('SSIM')
plt.xlabel('Epochs')
plt.legend(['Train','Val'],loc='upper right')
plt.show()
```
